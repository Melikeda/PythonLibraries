```python
from google.colab import files # Google Colab’ın files modülünü yüklüyorsun. Bu modül, bilgisayarından dosya yüklemeni sağlar.
uploaded = files.upload()  # Bu satır çalıştırıldığında bir dosya seçme penceresi açılır. Buradan bilgisayarındaki CSV dosyasını seçersin.

import pandas as pd
import io  # Python’un I/O (input/output) modülünü yüklüyorsun.
df = pd.read_csv(io.BytesIO(uploaded['train.csv']))
df.head()  # ilk 5 satırı göster
```

**5️. Yeni Kolonlar Eklemek (Feature Engineering)**

**1. Sabit Değerli Yeni Kolon**

```python
df['NewColumn'] = 1  # Tüm satırlara 1 değerini ekle
df.head()
```

**2. Mevcut Kolondan Hesaplama**

```python
df['AgePlusFare'] = df['Age'] + df['Fare']  # Yaş ve ücret toplamını yeni kolona ekle
df.head()
```

**3. Koşula Bağlı Yeni Kolon**

```python
df['Child'] = df['Age'] < 18  # 18 yaş altı yolcular Child = True
```

```python
df['HighFare'] = df['Fare'] > 50  # Ücret 50’den büyükse HighFare = True
df.head()
```

**4. Kategorik Kolonlardan Yeni Özellik**

```python
# Ücret aralıklarına göre kategori oluştur
df['FareCategory'] = pd.cut(
    df['Fare'],
    bins=[0,20,50,100,600],
    labels=['Cheap','Normal','Expensive','Luxury']
)
df.head()
```

```python
# Yaşa göre kategorik kolon
df['AgeCategory'] = pd.cut(
    df['Age'],
    bins=[0,12,18,60,100],
    labels=['Child','Teen','Adult','Senior']
)
df.head(10)
```

**5. String İşlemleri ile Yeni Kolon**

```python
df['Initial'] = df['Name'].str[0]  # İsmin baş harfini al
df.head()
```

```python
df['SexNum'] = df['Sex'].map({'male':0, 'female':1})  # Cinsiyet kolonunu 0-1 sayısal hale çevir
df.head()
```

**6. Gruplama ve Transform ile Kolon Eklemek**

```python
# Her sınıfın yaş ortalamasını yeni kolona ekle
df['AvgAgeByClass'] = df.groupby('Pclass')['Age'].transform('mean')
```

```python
# Her sınıfın toplam ücretini ekle
df['TotalFareByClass'] = df.groupby('Pclass')['Fare'].transform('sum')
df.head()
```

**7. Koşullu Fonksiyonlarla Yeni Kolon**

```python
# Fonksiyon tanımla
def age_group(age):
    if age < 18:
        return 'Child'
    elif age < 60:
        return 'Adult'
    else:
        return 'Senior'

# Yeni kolonu oluştur
df['AgeGroup'] = df['Age'].apply(age_group)
df
```
