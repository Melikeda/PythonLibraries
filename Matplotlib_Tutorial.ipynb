```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

**Basic Graph**

```python
x = [0,1,2,3]
y = [0,2,4,6]

plt.plot(x,y)
plt.title('Hi Graph')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

```python
x = [0,1,2,3]
y = [0,2,4,6]

plt.plot(x,y)
plt.title('Hi Graph')
plt.xlabel('X')
plt.ylabel('Y')

plt.xticks([0,1,2,3])
plt.yticks([0,2,4,6,8,10])

plt.show()
```

```python
x = [0,1,2,3]
y = [0,2,4,6]

plt.plot(x,y, linewidth = 2, linestyle='--', marker = '.', markersize =10,markeredgecolor = 'blue')

plt.plot(x,y)
plt.title('Hi Graph')
plt.xlabel('X')
plt.ylabel('Y')

plt.xticks([0,1,2,3])
plt.yticks([0,2,4,6,8,10])

plt.savefig('myHiGraph.png', dpi= 300)

plt.show()
```

**Bar Chart**

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

bars = plt.bar(labels,values)
bars[0].set_hatch('/')
bars[1].set_hatch('*')
bars[2].set_hatch('o')

plt.figure(figsize=(6,4))

plt.show()
```
