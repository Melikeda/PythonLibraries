```python
import seaborn as sns
import pandas as pd
```

**Grouping and Aggregation**

```python
df = sns.load_dataset('tips')
df
```

```python
df.groupby('total_bill')['TIP'].mean()
```

```python
df.groupby('day')['TIP'].agg(['mean','max','min'])
```

```python
df.groupby('total_bill').agg({
    'size':'mean',
    'TIP':'sum',
})
```

```python
df.groupby('day').size()
```
