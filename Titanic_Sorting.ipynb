```python
from google.colab import files # Google Colab’ın files modülünü yüklüyorsun. Bu modül, bilgisayarından dosya yüklemeni sağlar.
uploaded = files.upload()  # Bu satır çalıştırıldığında bir dosya seçme penceresi açılır. Buradan bilgisayarındaki CSV dosyasını seçersin.

import pandas as pd
import io  # Python’un I/O (input/output) modülünü yüklüyorsun.
df = pd.read_csv(io.BytesIO(uploaded['train.csv']))
df.head()  # ilk 5 satırı göster
```

**6️. Sıralama (Sorting)**

**1. Tek Sütuna Göre Sıralama**

```python
df_sorted = df.sort_values(by='Age')  # Age sütununa göre küçükten büyüğe sıralama
df
```

```python
df_sorted_desc = df.sort_values(by='Age', ascending=False)  # Yaş sütununa göre büyükten küçüğe
```

**2. Birden Fazla Sütuna Göre Sıralama**

```python
# Pclass küçükten büyüğe, Age büyükten küçüğe
df_sorted_multi = df.sort_values(by=['Pclass','Age'], ascending=[True, False])
df.head()
```

**3. Satırları İndeksine Göre Sıralama**

```python
# İndeks değerine göre sıralama
df_sorted_index = df.sort_index()
df
```

**4. Yeni DataFrame Olarak Saklamak veya Atama Yapmak**

```python
df = df.sort_values(by='Fare')  # Orijinal df’yi güncelle
df
```
