```python
import seaborn as sns
import pandas as pd
```

**Feature Engineering**

```python
df = sns.load_dataset('tips')
df
```

```python
df['NewColumn'] = "Hello"
df.head()
```

```python
df['New_Column'] = "Bro"
df.head()
```

```python
df['say'] = df['NewColumn'] + df['New_Column']
df.head()
```

```python
df['LitteSize'] = df['size'] < 3
df.head()
```

```python
df['sizeCategory'] = pd.cut(
    df['size'],
    bins=[0,2,3,4,5],
    labels=['Cheap','Normal','Expensive','Luxury']
)
df
```

```python
df['InitialSex'] = df['sex'].str[0]
df.head()
```

```python
df['InitialSex'] = df['sex'].str[0:3]
df.head()
```
