```python
import seaborn as sns
import pandas as pd
```

**Data Visualization**

```python
df = sns.load_dataset('tips')
df
```

```python
df.plot()  # Tüm grafik türlerini çağırabileceğin ana fonksiyondur.
```

```python
 # Çizgi Grafiği (Line Plot) Zaman serileri için çok kullanılır.

df.plot.line() # or "df.plot(kind='line')"
```

```python
# Sütun Grafiği (Bar Plot) Kategori karşılaştırmaları için kullanılır.

df.plot.bar()  # or "df.plot(kind='bar')"
```

```python
# Yatay bar grafiği

df.plot.barh()
```

```python
# Histogram (Histogram Plot) Veri dağılımını gösterir.

df.plot.hist()   # or "df['column'].plot.hist()"
```

```python
#Kutu Grafiği (Box Plot)  Aykırı değerleri ve dağılımı görürsün.

df.plot.box()  # or "df.boxplot()"
```

```python
# Alan Grafiği (Area Plot)  Zaman içinde değerlerin toplam etkisini gösterir.

df.plot.area()
```

```python
# Serpme Grafiği (Scatter Plot)  İki değişken arasındaki ilişkiyi anlamak için.

df.plot.scatter(x='day', y='time')
```

```python
# Pasta Grafiği (Pie Chart)  Kategori oranlarını göstermek için.

df['size'].plot.pie()
```

```python
# Yoğunluk Grafiği (Density / KDE Plot)  Dağılımın yoğunluk eğrisini çizer.

df.plot.kde()   # or "df.plot.density()"
```

```python
# Hexbin Grafiği  Büyük veri setlerinde 2 değişken ilişkisi için.

df.plot.hexbin(x='size', y='tip', gridsize=25)
```
