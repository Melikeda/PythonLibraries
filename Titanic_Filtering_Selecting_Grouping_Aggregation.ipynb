```python
from google.colab import files # Google Colab’ın files modülünü yüklüyorsun. Bu modül, bilgisayarından dosya yüklemeni sağlar.
uploaded = files.upload()  # Bu satır çalıştırıldığında bir dosya seçme penceresi açılır. Buradan bilgisayarındaki CSV dosyasını seçersin.

import pandas as pd
import io  # Python’un I/O (input/output) modülünü yüklüyorsun.
df = pd.read_csv(io.BytesIO(uploaded['train.csv']))
df.head()  # ilk 5 satırı göster
```

**3️. Filtreleme ve Seçim (Filtering / Selecting)**

**1. Tek Sütun Seçmek**

```python
df['Age']          # Tek sütun: Series döner
```

```python
df[['Age']]        # Tek sütun: DataFrame döner
```

```python
df[['Age','Fare']] # Birden fazla sütun döner
```

**2. Satır Seçmek (Index ile)**

```python
df.iloc[0]   # İlk satır
```

```python
df.iloc[0:5] # İlk 5 satır
```

```python
df.iloc[:,0] # Tüm satırların ilk sütunu
```

**3. Satır ve Sütun Seçmek**

```python
df.loc[0,'Age']      # 0. satır ve 'Age' sütunu
```

```python
df.loc[0:5,['Age','Fare']]  # Satır 0-5 ve belirli sütunlar
```

**4. Koşullu Filtreleme**

```python
df[df['Age'] > 30]  # Age > 30
```

```python
df[(df['Age'] > 30) & (df['Sex'] == 'female')]   # Age > 30 ve Sex = 'female'
```

```python
df[(df['Age'] > 30) | (df['Fare'] > 50)]  # Age > 30 veya Fare > 50
```

**5. Belirli Değerleri Filtreleme**

```python
df[df['Pclass'].isin([1])]  # Pclass 1 olan yolcular
```

```python
df[df['Pclass'].isin([1,2])]  # Birden fazla sınıf
```

**6. Metin Bazlı Filtreleme (String Filtering)**

```python
df[df['Name'].str.contains('John')]  # İsmi “John” içeren yolcular
```

```python
df[df['Sex'].str.lower() == 'female']   # Cinsiyeti 'female' olan
```

**7. Sıralı ve Koşullu Filtreleme**

```python
df[df['Age'] > 30].sort_values(by='Age', ascending=True)  # Yaş > 30 olanları yaşa göre sırala
```

**8. Boolean Mask Kullanmak**

```python
mask = df['Fare'] > 50
df[mask]
```

**9. Sütun ve Satırları Birlikte Seçmek**

```python
df.loc[df['Age'] > 30, ['Name','Age','Fare']]
```

**10. Negatif Filtreleme**

```python
df[~(df['Age'] > 30)]  # Age <= 30 olanlar
```

**4️. Gruplama ve Özetleme (Grouping / Aggregation)**

**1. Basit**

```python
df.groupby('Sex')['Survived'].mean()  # Cinsiyete göre hayatta kalma ortalaması
```

**2. Birden Fazla Sütuna Göre Gruplama**

```python
df.groupby(['Sex','Pclass'])['Age'].mean()  # Cinsiyet ve sınıfa göre yaş ortalaması
```

**3. Birden Fazla İstatistik**

```python
df.groupby('Pclass')['Age'].agg(['mean','max','min'])  # Yaş için ortalama, maksimum ve minimum
```

**4. Farklı Sütunlar için Farklı Özetler**

```python
df.groupby('Pclass').agg({
    'Age':'mean',         # Yaş ortalaması
    'Fare':'sum',         # Toplam ücret
    'Survived':'mean'     # Hayatta kalma oranı
})
```

**5. Gruplama Sonrası Filtreleme**

```python
df.groupby('Pclass').filter(lambda x: x['Age'].mean() > 30)  # Sadece ortalama yaşı 30’dan büyük olan sınıfları al
```

**6. Gruplama Sonrası Dönüştürme**

```python
df['AvgAgeByClass'] = df.groupby('Pclass')['Age'].transform('mean')  # Her yolcuya sınıfın yaş ortalamasını ekle
```

**7. Gruplama Sonrası Uygulama**

```python
df.groupby('Pclass').apply(lambda x: x[x['Fare'] == x['Fare'].max()])  # Her gruptaki en yüksek ücreti alan yolcuyu getir
```

**8. Boyutları ve Sayıları Görmek**

```python
df.groupby('Pclass').size()  # Her sınıfta kaç yolcu var
```

```python
df.groupby('Pclass')['Survived'].sum() # Her sınıfta kaç kişi hayatta kalmış
```
