**Veri Temizleme (Data Cleaning)**

**Veri Temizleme İçin Kullanılan Pandas Methodları**

**1. Eksik Değerleri Kontrol Etmek**

```python
df.isnull()  # Her hücrede eksik değer varsa True döner
```

```python
df.isnull().sum() # Her sütunda kaç eksik değer var
```

```python
df.notnull()  # Eksik olmayan hücreleri True döner
```

**2. Eksik Değerleri Doldurmak (Fillna)**

```python
print(df.columns)
```

```python
df['Age'] = df['Age'].fillna(0)        # Tüm boşları 0 ile doldur
df
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())  # Ortalama ile doldur
df
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])  # En sık görülen ile doldur
df
```

**3️. Eksik Değerleri Silmek (Dropna)**

```python
df = df.dropna(axis=0)  # # Boş değer içeren satırları sil
df
```

```python
df = df.dropna(axis=1)   # Boş değer içeren sütunları sil
df
```

**4️. Veri Tipini Düzeltmek (astype)**

```python
df['Survived'] = df['Survived'].astype(int)  # Float → int
```

```python
df['Pclass'] = df['Pclass'].astype(str)      # Sayı → string
```

**5️. Gereksiz Sütunları Kaldırmak (drop)**

axis=0 → satır silmek

axis=1 → sütun silmek

```python
df = df.drop(['Cabin', 'Ticket'], axis=1)   # Sütunları sil
df
```

**6️. Sütun veya Satır İsmini Değiştirmek (rename)**

```python
df = df.rename(columns={'Pclass':'Class', 'Sex':'Gender'})
```

**7️. Tekrarlayan Satırları Kaldırmak (drop_duplicates)**

```python
df = df.drop_duplicates()
```

**8️. String Veriyi Temizlemek**

```python
df['Name'] = df['Name'].str.strip()      # Baş ve sondaki boşlukları sil
df
```

```python
df.head()
```

```python
df['Gender'] = df['Gender'].str.upper()        # Büyük harfe çevir
df.head()
```

```python
df['Gender'] = df['Gender'].str.lower()        # Küçük harfe çevir
df.head()
```

**9️. Koşullu Düzenlemeler (loc)**

```python
df.loc[df['Age'] < 0, 'Age'] = df['Age'].mean()  # # Age sütunu 0’dan küçükse ortalama ile değiştir
```

**11️. Boolean Kolon Eklemek**

```python
df['Child'] = df['Age'] < 18   # 18 yaş altı yolcular True
df.head()
```
