```python
import seaborn as sns
import pandas as pd
```

**Data Cleaning**

```python
df = sns.load_dataset('tips')
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.notnull()
```

```python
df = df.rename(columns={'tip':'TIP', 'Sex':'Gender'})
df.columns
```

```python
df['sex'] = df['sex'].str.upper()
df.head()
```

```python
df['day'] = df['day'].str.lower()
df.head()
```

```python
df.loc[df['TIP'] < 2, 'TIP'] = df['TIP'].mean()
df
```
